```python
from algos.ratio.live_trader import liveTrading
import pandas as pd
```

```python
lt = liveTrading('TROYBTC')
```

```python
trades = lt.exchange.sapi_get_margin_mytrades({'symbol': lt.symbol, 'isIsolated': 'TRUE'})
```

```python
trades_df = pd.DataFrame(trades)
```

```python
trades_df['price'] = trades_df['price'].astype(float)
```

```python
trades_df['qty'] = trades_df['qty'].astype(float)
```

```python
reverse_df = trades_df[::-1].reset_index(drop=True)
```

```python
select_df = pd.DataFrame()
for idx, row in reverse_df.iterrows():
    if row['isBuyer'] != reverse_df.iloc[0]['isBuyer']:
        break
    
    select_df = select_df.append(row, ignore_index=True)
```

```python
select_df['price'] * select_df['qty']
```
