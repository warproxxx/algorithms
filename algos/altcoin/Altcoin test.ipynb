```python
from backtest import get_df, get_hma, tradingStrategy, get_params_lines, CommInfoFractional, plot
import pandas as pd
import backtrader as bt
```

```python
price_df = get_df('BTC-PERP')
```

```python
price_df['price_hall'] = get_hma(price_df['close'], 30)
price_df['volatility'] = (price_df['close'].std()/price_df['close']) * 100
```

```python
initial_cash = 1000
mult = 1
params = {'stop': 800}
pair = 'BTC-PERP'
```

```python
PandasData_Custom = get_params_lines(price_df.columns)
```

```python
price_data = PandasData_Custom(dataname=price_df)

cerebro = bt.Cerebro()

cerebro.adddata(price_data, name='data')
cerebro.addstrategy(tradingStrategy, parameters=params)
cerebro.addanalyzer(bt.analyzers.SharpeRatio, riskfreerate=0.0, annualize=True, timeframe=bt.TimeFrame.Days)
cerebro.addanalyzer(bt.analyzers.TradeAnalyzer)

cerebro.broker = bt.brokers.BackBroker(cash=initial_cash, slip_perc=0.01/100, commission = CommInfoFractional(commission=(0.075*mult)/100, mult=mult), slip_open=True, slip_out=True)  # 0.5%
run = cerebro.run()
```

```python
portfolio, trades, operations, stops_triggered = run[0].get_logs()
```
